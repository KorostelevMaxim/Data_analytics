Делаем сводную таблицу и индексом rooms, чтобы при наличии пропусков площадей кухни и жилого помещения, была возможность заменить эти пропуски на медианные значения в соответствии с количеством комнат

```python
df_area = df.pivot_table(index = 'rooms', values = ['living_area','kitchen_area','total_area'], aggfunc='median')
```

```python
df_area
```

```python
df_area.loc[0]['kitchen_area'] = 0
for room in df_area.index:
    df.loc[(df['rooms'] == room) & (df['kitchen_area'].isna()), 'kitchen_area'] = df_area.kitchen_area[room]
    df.loc[(df['rooms'] == room) & (df['living_area'].isna()), 'living_area'] = df_area.living_area[room]        
```

```python
df.isna().sum()
```

Делаем сводную таблицу и индексом total_floors, чтобы при наличии пропусков высоты потолков, была возможность заменить эти пропуски на медианные значения в соответствии с количеством этажей

```python
df_ceiling = df.pivot_table(index = 'floors_total', values ='ceiling_height', aggfunc='median')
```

```python
df_ceiling
```

```python
for floor in df_ceiling.index:
    df.loc[(df['floors_total'] == floor) & (df['ceiling_height'].isna()), 'ceiling_height'] = df_ceiling.ceiling_height[floor]
```

```python
df.isna().sum()
```

### Выведем клоличество уникальных названий населёных пунктов после обработки дубликатов

```python
len(df['locality_name'].sort_values().unique())
```

```python
def locality_edit(name):
    try:
        name = name.lower()                           
        if 'поселок городского типа' in name:
            loc_name = 'поселок городского типа'
            return 'поселок' + name[len(loc_name):]
        elif 'посёлок городского типа' in name:
            loc_name = 'посёлок городского типа'
            return 'поселок' + name[len(loc_name):]
        elif 'коттеджный посёлок' in name:
            loc_name = 'коттеджный посёлок'
            return 'поселок' + name[len(loc_name):]
        elif 'коттеджный поселок' in name:
            loc_name = 'коттеджный поселок'
            return 'поселок' + name[len(loc_name):]
        elif 'городской поселок' in name:
            loc_name = 'городской поселок'
            return 'поселок' + name[len(loc_name):]
        elif 'городской посёлок' in name:
            loc_name = 'городской посёлок'
            return 'поселок' + name[len(loc_name):]
        elif 'посёлок' in name:
            loc_name = 'посёлок'
            return 'поселок' + name[len(loc_name):]
        elif 'поселок' in name:
            loc_name = 'поселок'
            return 'поселок' + name[len(loc_name):]
        elif 'село' in name:
            loc_name = 'село'
            return 'поселок' + name[len(loc_name):]
        else:
            return name
    except:
        return 'нет названия'
```

```python
df['locality_name']= df['locality_name'].apply(locality_edit)
```
